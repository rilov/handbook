# Multiple Linear Regression - House Price Prediction

## 🎯 Goal
Predict house prices using **multiple features**:
- Size (square feet)
- Number of bedrooms
- Age of house
- Location score (1-10)

## 📚 What You'll Learn
- How to work with multiple features
- How to interpret coefficients for each feature
- How features interact to determine price
- How to make predictions with multiple inputs

## Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Set random seed
np.random.seed(42)

# Set plot style
sns.set_style('whitegrid')

print("✅ Libraries imported successfully!")
```

## Step 2: Create Sample Data

Create a dataset with multiple features.

```python
# Create dataset
data = {
    'Size (sq ft)': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 2200, 2800, 3200],
    'Bedrooms': [2, 3, 3, 4, 4, 5, 5, 3, 4, 4],
    'Age (years)': [10, 5, 2, 15, 1, 8, 3, 7, 12, 4],
    'Location (1-10)': [5, 7, 8, 6, 9, 8, 10, 7, 6, 9],
    'Price ($1000s)': [200, 300, 400, 450, 600, 650, 800, 420, 520, 680]
}

df = pd.DataFrame(data)

print("🏠 House Price Dataset (Multiple Features)")
print("=" * 70)
print(df)
print(f"\nDataset shape: {df.shape[0]} houses, {df.shape[1]-1} features")
```

## Step 3: Explore the Data

Let's understand our features better.

```python
print("📊 Dataset Statistics")
print("=" * 70)
print(df.describe())

print("\n📈 Feature Ranges:")
for col in df.columns[:-1]:  # Exclude price
    print(f"  {col:20s}: {df[col].min():6.0f} - {df[col].max():6.0f}")
```

## Step 4: Visualize Relationships

Plot each feature vs price to see individual relationships.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
features = ['Size (sq ft)', 'Bedrooms', 'Age (years)', 'Location (1-10)']

for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    
    axes[row, col].scatter(df[feature], df['Price ($1000s)'], 
                          s=100, alpha=0.6, edgecolors='black')
    axes[row, col].set_xlabel(feature, fontsize=11)
    axes[row, col].set_ylabel('Price ($1000s)', fontsize=11)
    axes[row, col].set_title(f'Price vs {feature}', fontsize=12, fontweight='bold')
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ Visualizations complete!")
print("   - Size, Bedrooms, Location: Positive relationship (↑ feature → ↑ price)")
print("   - Age: Negative relationship (↑ age → ↓ price)")
```

## Step 5: Check Feature Correlations

See how features relate to each other and to price.

```python
# Calculate correlation matrix
corr_matrix = df.corr()

# Visualize
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, fmt='.2f', cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("📊 Correlation with Price:")
print("=" * 40)
price_corr = corr_matrix['Price ($1000s)'].sort_values(ascending=False)
for feature, corr in price_corr.items():
    if feature != 'Price ($1000s)':
        print(f"  {feature:20s}: {corr:6.3f}")

print("\n💡 Interpretation:")
print("   Values close to +1: Strong positive correlation")
print("   Values close to -1: Strong negative correlation")
print("   Values close to  0: Weak/no correlation")
```

## Step 6: Prepare Data for Training

Split features (X) and target (y).

```python
# Features (X) and target (y)
X = df[['Size (sq ft)', 'Bedrooms', 'Age (years)', 'Location (1-10)']]
y = df['Price ($1000s)']

print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape:   {y.shape}")
print("\n✅ Data prepared for training!")
```

## Step 7: Train the Model

Create and train a Multiple Linear Regression model.

```python
# Create and train model
model = LinearRegression()
model.fit(X, y)

print("🤖 Model Training Complete!")
print("=" * 70)
print("✅ Model has learned how all features affect price")
```

## Step 8: Understand Model Parameters

Each feature has its own coefficient!

```python
intercept = model.intercept_
coefficients = model.coef_

print("📋 Model Parameters")
print("=" * 70)
print(f"Intercept (β₀): ${intercept:.2f}k\n")

print("Coefficients:")
for feature, coef in zip(X.columns, coefficients):
    print(f"  {feature:20s} (β): ${coef:8.4f}k")

print("\n📐 Equation:")
print(f"Price = {intercept:.2f}", end="")
for feature, coef in zip(X.columns, coefficients):
    sign = "+" if coef >= 0 else "-"
    print(f" {sign} {abs(coef):.4f}×{feature}", end="")
print()

print("\n💡 What Each Coefficient Means:")
print("=" * 70)
for feature, coef in zip(X.columns, coefficients):
    if coef > 0:
        print(f"  ✅ {feature:20s}: Each unit increase adds ${abs(coef):.2f}k")
    else:
        print(f"  ❌ {feature:20s}: Each unit increase subtracts ${abs(coef):.2f}k")

print("\n📝 Examples:")
print(f"  - Adding 1000 sq ft adds: ${coefficients[0]*1000:.0f}k")
print(f"  - Adding 1 bedroom adds: ${coefficients[1]:.0f}k")
print(f"  - Each year older subtracts: ${abs(coefficients[2]):.0f}k")
print(f"  - Improving location by 1 point adds: ${coefficients[3]:.0f}k")
```

## Step 9: Make Predictions

Predict prices for our training data.

```python
# Predict
y_pred = model.predict(X)

# Create results DataFrame
results = df.copy()
results['Predicted Price'] = y_pred
results['Error'] = results['Price ($1000s)'] - results['Predicted Price']

print("🔮 Predictions")
print("=" * 90)
print(results.to_string(index=False))

print("\n📊 Error Analysis:")
print(f"  Mean error: ${results['Error'].mean():.2f}k")
print(f"  Max error:  ${results['Error'].abs().max():.2f}k")
```

## Step 10: Predict for New Houses

Let's predict prices for houses not in our dataset!

```python
# New houses
new_houses = pd.DataFrame({
    'Size (sq ft)': [2200, 3500, 1800],
    'Bedrooms': [3, 4, 2],
    'Age (years)': [5, 10, 1],
    'Location (1-10)': [7, 9, 6]
})

# Predict
new_predictions = model.predict(new_houses)

print("🏠 Predictions for New Houses")
print("=" * 70)

for i in range(len(new_houses)):
    print(f"\nHouse {i+1}:")
    print(f"  Size: {new_houses.iloc[i]['Size (sq ft)']} sq ft")
    print(f"  Bedrooms: {new_houses.iloc[i]['Bedrooms']}")
    print(f"  Age: {new_houses.iloc[i]['Age (years)']} years")
    print(f"  Location: {new_houses.iloc[i]['Location (1-10)']}/10")
    print(f"  💰 Predicted Price: ${new_predictions[i]:.0f}k (${new_predictions[i]*1000:,.0f})")
```

## Step 11: Evaluate Model Performance

```python
# Calculate metrics
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)

print("📊 Model Performance")
print("=" * 40)
print(f"R² Score:                {r2:.4f}")
print(f"Mean Absolute Error:     ${mae:.2f}k")
print(f"Root Mean Squared Error: ${rmse:.2f}k")

print("\n💡 Interpretation:")
print(f"  - R² = {r2:.1%}: Model explains {r2:.1%} of price variance")
print(f"  - MAE = ${mae:.0f}k: Predictions are off by ${mae:.0f}k on average")

if r2 > 0.95:
    print("  - ✅ Excellent fit!")
elif r2 > 0.85:
    print("  - ✅ Very good fit!")
elif r2 > 0.70:
    print("  - ✅ Good fit!")
else:
    print("  - ⚠️  Model could be improved")
```

## Step 12: Visualize Predictions vs Actual

```python
plt.figure(figsize=(10, 6))

# Plot perfect prediction line
min_val = min(y.min(), y_pred.min())
max_val = max(y.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 
         'r--', linewidth=2, label='Perfect prediction', alpha=0.7)

# Plot actual vs predicted
plt.scatter(y, y_pred, s=100, alpha=0.6, edgecolors='black')

plt.xlabel('Actual Price ($1000s)', fontsize=12)
plt.ylabel('Predicted Price ($1000s)', fontsize=12)
plt.title('Actual vs Predicted Prices', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ If points are close to the red line, predictions are accurate!")
```

## Step 13: Feature Importance Visualization

```python
# Create feature importance plot
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': coefficients
}).sort_values('Coefficient', key=abs, ascending=False)

plt.figure(figsize=(10, 6))
colors = ['green' if x > 0 else 'red' for x in feature_importance['Coefficient']]
plt.barh(feature_importance['Feature'], feature_importance['Coefficient'], color=colors, alpha=0.7)
plt.xlabel('Coefficient Value', fontsize=12)
plt.title('Feature Importance (Coefficient Magnitude)', fontsize=14, fontweight='bold')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

print("📊 Feature Importance:")
print("  Green bars: Positive effect (increases price)")
print("  Red bars: Negative effect (decreases price)")
print("  Longer bars: Stronger effect")
```

## Step 14: Interactive Prediction Function

```python
def predict_price(size, bedrooms, age, location):
    """
    Predict house price based on features.
    
    Parameters:
    size (float): House size in square feet
    bedrooms (int): Number of bedrooms
    age (float): Age of house in years
    location (int): Location score (1-10)
    
    Returns:
    float: Predicted price in thousands of dollars
    """
    # Create input array
    features = np.array([[size, bedrooms, age, location]])
    
    # Predict
    price = model.predict(features)[0]
    
    print("\n🏠 House Features:")
    print("=" * 40)
    print(f"  Size:     {size:,} sq ft")
    print(f"  Bedrooms: {bedrooms}")
    print(f"  Age:      {age} years")
    print(f"  Location: {location}/10")
    
    print(f"\n💰 Predicted Price: ${price:.0f}k (${price*1000:,.0f})")
    
    print("\n📐 Calculation Breakdown:")
    print(f"  Base price:        ${intercept:.2f}k")
    print(f"  + Size effect:     ${coefficients[0]*size:.2f}k")
    print(f"  + Bedrooms effect: ${coefficients[1]*bedrooms:.2f}k")
    print(f"  - Age effect:      ${coefficients[2]*age:.2f}k")
    print(f"  + Location effect: ${coefficients[3]*location:.2f}k")
    print(f"  " + "=" * 35)
    print(f"  Total:             ${price:.2f}k")
    
    return price

# Try it!
print("🎯 Example Predictions:")
print("=" * 70)

predict_price(size=2500, bedrooms=3, age=8, location=7)
predict_price(size=3500, bedrooms=4, age=2, location=9)
predict_price(size=1500, bedrooms=2, age=20, location=5)
```

### Key Differences from Simple Linear Regression:
- **Multiple features**: Each has its own coefficient
- **Interpretation**: "Holding other features constant"
- **Complexity**: More powerful but harder to visualize
- **Multicollinearity**: Need to check feature correlations

### Important Concepts:
- **Positive coefficient**: Feature increases price
- **Negative coefficient**: Feature decreases price
- **Magnitude**: Larger absolute value = stronger effect
- **R² Score**: Still measures overall fit quality
