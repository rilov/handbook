# Simple Linear Regression - House Price Prediction

## 🎯 Goal
Predict house prices based on size (square feet)

## 📚 What You'll Learn
- How to create and visualize data
- How to train a simple linear regression model
- How to interpret coefficients (slope and intercept)
- How to make predictions
- How to evaluate model performance

## Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Set random seed for reproducibility
np.random.seed(42)

print("✅ Libraries imported successfully!")
```

## Step 2: Create Sample Data

We'll create a dataset of houses with their sizes and prices.

```python
# House sizes (in square feet)
sizes = np.array([1000, 1500, 2000, 2500, 3000, 3500, 4000])

# House prices (in $1000s)
prices = np.array([200, 300, 400, 500, 600, 700, 800])

# Create DataFrame
df = pd.DataFrame({
    'Size (sq ft)': sizes,
    'Price ($1000s)': prices
})

print("🏠 House Price Dataset")
print("=" * 40)
print(df)
print(f"\nNumber of houses: {len(df)}")
print(f"Size range: {sizes.min()} - {sizes.max()} sq ft")
print(f"Price range: ${prices.min()}k - ${prices.max()}k")
```

## Step 3: Visualize the Data

Let's plot the data to see if there's a linear relationship.

```python
plt.figure(figsize=(10, 6))
plt.scatter(sizes, prices, color='blue', s=100, alpha=0.7, edgecolors='black')
plt.xlabel('House Size (sq ft)', fontsize=12)
plt.ylabel('Price ($1000s)', fontsize=12)
plt.title('House Size vs Price', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Can you see the linear pattern? Points form a straight line!")
```

## Step 4: Prepare Data for Training

Reshape data for sklearn (needs 2D array for X)

```python
# Reshape X to 2D array (required by sklearn)
X = sizes.reshape(-1, 1)
y = prices

print(f"X shape: {X.shape}  (samples, features)")
print(f"y shape: {y.shape}  (samples,)")
print("\n✅ Data prepared for training!")
```

## Step 5: Train the Model

Create and train a Linear Regression model.

```python
# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

print("🤖 Model Training Complete!")
print("=" * 40)
print("✅ Model has learned the pattern from the data")
```

## Step 6: Understand the Model Parameters

The model learned two parameters:
- **Slope (β₁)**: How much price changes per square foot
- **Intercept (β₀)**: Base price when size = 0

```python
slope = model.coef_[0]
intercept = model.intercept_

print("📋 Model Parameters")
print("=" * 40)
print(f"Slope (β₁):     ${slope:.4f}k per sq ft")
print(f"Intercept (β₀): ${intercept:.2f}k")
print(f"\n📐 Equation: Price = {intercept:.2f} + {slope:.4f} × Size")

print("\n💡 What This Means:")
print(f"   - Base price (land, permits): ${intercept:.0f}k")
print(f"   - Each additional sq ft adds: ${slope*1000:.0f}")
print(f"   - A 1000 sq ft increase adds: ${slope*1000:.0f}k")
```

## Step 7: Make Predictions

Use the trained model to predict prices.

```python
# Predict on training data
y_pred = model.predict(X)

print("🔮 Predictions on Training Data")
print("=" * 60)
print(f"{'Size (sq ft)':<15} {'Actual Price':<15} {'Predicted':<15} {'Error':<10}")
print("-" * 60)

for i in range(len(X)):
    actual = y[i]
    predicted = y_pred[i]
    error = actual - predicted
    print(f"{X[i][0]:<15.0f} ${actual:<14.0f}k ${predicted:<14.2f}k ${error:<9.2f}k")

print("\n✅ Perfect predictions! (Because we're using training data)")
```

## Step 8: Predict for New Houses

Let's predict prices for houses we haven't seen before!

```python
# New house sizes
new_sizes = np.array([1800, 2800, 3800]).reshape(-1, 1)
new_predictions = model.predict(new_sizes)

print("🏠 Predictions for New Houses")
print("=" * 40)

for i in range(len(new_sizes)):
    size = new_sizes[i][0]
    price = new_predictions[i]
    print(f"  {size:.0f} sq ft → ${price:.0f}k (${price*1000:.0f})")

print("\n💡 Try your own! Modify the new_sizes array above.")
```

## Step 9: Evaluate Model Performance

How well does our model fit the data?

```python
# Calculate metrics
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)

print("📊 Model Performance Metrics")
print("=" * 40)
print(f"R² Score:              {r2:.4f}")
print(f"Mean Squared Error:    {mse:.2f}")
print(f"Root Mean Squared Error: ${rmse:.2f}k")
print(f"Mean Absolute Error:   ${mae:.2f}k")

print("\n💡 Interpretation:")
print(f"   - R² = {r2:.1%} means the model explains {r2:.1%} of variance")
if r2 > 0.99:
    print("   - ✅ Excellent fit! (R² > 0.99)")
elif r2 > 0.90:
    print("   - ✅ Very good fit! (R² > 0.90)")
elif r2 > 0.70:
    print("   - ✅ Good fit! (R² > 0.70)")
else:
    print("   - ⚠️  Model could be improved (R² < 0.70)")

print(f"   - MAE = ${mae:.0f}k means predictions are off by ${mae:.0f}k on average")
```

## Step 10: Visualize the Regression Line

Plot the data points and the fitted line.

```python
plt.figure(figsize=(12, 6))

# Plot actual data
plt.scatter(sizes, prices, color='blue', s=100, alpha=0.7, 
           edgecolors='black', label='Actual data', zorder=3)

# Plot regression line
plt.plot(sizes, y_pred, color='red', linewidth=2, 
        label=f'Regression line: y = {intercept:.2f} + {slope:.4f}x', zorder=2)

# Add prediction examples
plt.scatter(new_sizes, new_predictions, color='green', s=150, 
           marker='*', edgecolors='black', label='New predictions', zorder=4)

plt.xlabel('House Size (sq ft)', fontsize=12)
plt.ylabel('Price ($1000s)', fontsize=12)
plt.title('Linear Regression: House Size vs Price', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Regression line plotted!")
print("   - Blue dots: Training data")
print("   - Red line: Model's predictions")
print("   - Green stars: New predictions")
```

## Step 11: Interactive Prediction Function

Create a function to easily predict prices for any size.

```python
def predict_house_price(size):
    """
    Predict house price based on size.
    
    Parameters:
    size (float): House size in square feet
    
    Returns:
    float: Predicted price in thousands of dollars
    """
    price = model.predict([[size]])[0]
    
    print(f"\n🏠 House Size: {size:,.0f} sq ft")
    print(f"💰 Predicted Price: ${price:.0f}k (${price*1000:,.0f})")
    print(f"\n📐 Calculation:")
    print(f"   Price = {intercept:.2f} + {slope:.4f} × {size}")
    print(f"   Price = {intercept:.2f} + {slope*size:.2f}")
    print(f"   Price = ${price:.2f}k")
    
    return price

# Try it!
print("🎯 Try Different House Sizes:")
print("=" * 40)
predict_house_price(2200)
predict_house_price(3200)
predict_house_price(5000)
```
