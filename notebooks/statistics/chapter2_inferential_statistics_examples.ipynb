```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print("✅ Libraries imported successfully!")
```

## Load the Penguins Dataset

```python
# Load penguins dataset
penguins = sns.load_dataset('penguins')

# Remove missing values
penguins = penguins.dropna()

print("🐧 Dataset Shape:", penguins.shape)
print("\nFirst 5 rows:")
penguins.head()
```

```python
# Dataset overview
print("ℹ️ Dataset Information:\n")
penguins.info()
print("\n📊 Species Distribution:\n")
print(penguins['species'].value_counts())
print("\n🏝️ Island Distribution:\n")
print(penguins['island'].value_counts())
```

## Example 1: Understanding Population vs Sample

**Scenario:** We have data on 333 penguins. Let's pretend this is our **entire population**. We'll take samples and see how well they represent the population.

```python
print("="*60)
print("🐧 POPULATION vs SAMPLE")
print("="*60)
print()

# Population parameters (TRUE values)
population_mean = penguins['body_mass_g'].mean()
population_std = penguins['body_mass_g'].std()
population_size = len(penguins)

print("🌍 POPULATION (All 333 penguins):")
print(f"   Mean body mass: {population_mean:.2f} g")
print(f"   Std deviation: {population_std:.2f} g")
print(f"   Size: {population_size} penguins")
print()

# Take a random sample
sample_size = 30
sample = penguins['body_mass_g'].sample(n=sample_size, random_state=42)
sample_mean = sample.mean()
sample_std = sample.std()

print(f"📦 SAMPLE (Random {sample_size} penguins):")
print(f"   Mean body mass: {sample_mean:.2f} g")
print(f"   Std deviation: {sample_std:.2f} g")
print(f"   Size: {sample_size} penguins")
print()

print("🎯 HOW CLOSE IS OUR SAMPLE?")
print(f"   Difference in means: {abs(sample_mean - population_mean):.2f} g")
print(f"   Percentage error: {abs(sample_mean - population_mean) / population_mean * 100:.2f}%")
print()
print("="*60)
```

```python
# Visualize population vs sample
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Population distribution
axes[0].hist(penguins['body_mass_g'], bins=30, color='#4ECDC4', alpha=0.7, edgecolor='black')
axes[0].axvline(population_mean, color='red', linestyle='--', linewidth=2, 
                label=f'Population Mean: {population_mean:.0f}g')
axes[0].set_xlabel('Body Mass (g)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('POPULATION Distribution (n=333)', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

# Sample distribution
axes[1].hist(sample, bins=15, color='#FF6B6B', alpha=0.7, edgecolor='black')
axes[1].axvline(sample_mean, color='blue', linestyle='--', linewidth=2, 
                label=f'Sample Mean: {sample_mean:.0f}g')
axes[1].axvline(population_mean, color='red', linestyle='--', linewidth=2, 
                label=f'True Mean: {population_mean:.0f}g')
axes[1].set_xlabel('Body Mass (g)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_title(f'SAMPLE Distribution (n={sample_size})', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("💡 Notice: The sample mean is close to the population mean!")
print("   This is the power of random sampling.")
```

## Example 2: Central Limit Theorem (CLT)

**The Magic:** If you take many samples and calculate their means, those means will form a normal distribution!

```python
print("="*60)
print("🎯 CENTRAL LIMIT THEOREM IN ACTION")
print("="*60)
print()

# Take 1000 samples and calculate their means
num_samples = 1000
sample_size = 30
sample_means = []

for i in range(num_samples):
    sample = penguins['body_mass_g'].sample(n=sample_size)
    sample_means.append(sample.mean())

sample_means = np.array(sample_means)

print(f"📊 Took {num_samples} samples of size {sample_size}")
print(f"   Mean of sample means: {sample_means.mean():.2f} g")
print(f"   True population mean: {population_mean:.2f} g")
print(f"   Difference: {abs(sample_means.mean() - population_mean):.2f} g")
print()
print("="*60)
```

```python
# Visualize CLT
plt.figure(figsize=(12, 6))

# Plot distribution of sample means
plt.hist(sample_means, bins=50, color='#45B7D1', alpha=0.7, edgecolor='black', density=True)

# Overlay normal distribution
x = np.linspace(sample_means.min(), sample_means.max(), 100)
plt.plot(x, stats.norm.pdf(x, sample_means.mean(), sample_means.std()), 
         'r-', linewidth=3, label='Normal Distribution')

plt.axvline(population_mean, color='green', linestyle='--', linewidth=2, 
            label=f'True Mean: {population_mean:.0f}g')

plt.xlabel('Sample Mean Body Mass (g)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title(f'Distribution of {num_samples} Sample Means (n={sample_size})', 
          fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("🎉 CENTRAL LIMIT THEOREM:")
print("   ✅ Sample means form a normal distribution")
print("   ✅ Centered around the true population mean")
print("   ✅ This works even if the original data isn't normal!")
```

## Example 3: Confidence Intervals

**Question:** Based on a sample, what's the likely range for the true population mean?

```python
print("="*60)
print("📊 CONFIDENCE INTERVALS")
print("="*60)
print()

# Take a sample
sample = penguins['body_mass_g'].sample(n=50, random_state=123)
sample_mean = sample.mean()
sample_std = sample.std()
sample_size = len(sample)

# Calculate 95% confidence interval
confidence_level = 0.95
degrees_freedom = sample_size - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_freedom)

# Standard error
standard_error = sample_std / np.sqrt(sample_size)

# Margin of error
margin_of_error = t_critical * standard_error

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"📦 Sample Statistics (n={sample_size}):")
print(f"   Sample mean: {sample_mean:.2f} g")
print(f"   Sample std: {sample_std:.2f} g")
print(f"   Standard error: {standard_error:.2f} g")
print()
print(f"📏 95% Confidence Interval:")
print(f"   Lower bound: {ci_lower:.2f} g")
print(f"   Upper bound: {ci_upper:.2f} g")
print(f"   Margin of error: ±{margin_of_error:.2f} g")
print()
print(f"💡 Interpretation:")
print(f"   We are 95% confident that the true population mean")
print(f"   is between {ci_lower:.0f}g and {ci_upper:.0f}g")
print()
print(f"✅ True population mean: {population_mean:.2f} g")
if ci_lower <= population_mean <= ci_upper:
    print(f"   ✅ SUCCESS! Our interval captured the true mean!")
else:
    print(f"   ❌ MISS! Our interval didn't capture the true mean (happens 5% of the time)")
print()
print("="*60)
```

```python
# Visualize confidence interval
plt.figure(figsize=(12, 6))

# Plot sample distribution
plt.hist(sample, bins=20, color='#4ECDC4', alpha=0.7, edgecolor='black')

# Add lines
plt.axvline(sample_mean, color='blue', linestyle='-', linewidth=3, 
            label=f'Sample Mean: {sample_mean:.0f}g')
plt.axvline(ci_lower, color='red', linestyle='--', linewidth=2, 
            label=f'95% CI Lower: {ci_lower:.0f}g')
plt.axvline(ci_upper, color='red', linestyle='--', linewidth=2, 
            label=f'95% CI Upper: {ci_upper:.0f}g')
plt.axvline(population_mean, color='green', linestyle=':', linewidth=3, 
            label=f'True Mean: {population_mean:.0f}g')

# Shade confidence interval
plt.axvspan(ci_lower, ci_upper, alpha=0.2, color='red')

plt.xlabel('Body Mass (g)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('95% Confidence Interval for Population Mean', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Example 4: Effect of Sample Size on Confidence Intervals

**Question:** How does sample size affect the width of confidence intervals?

```python
# Calculate CIs for different sample sizes
sample_sizes = [10, 30, 50, 100, 200]
ci_results = []

for n in sample_sizes:
    sample = penguins['body_mass_g'].sample(n=n, random_state=42)
    mean = sample.mean()
    std = sample.std()
    se = std / np.sqrt(n)
    t_crit = stats.t.ppf(0.975, n-1)
    moe = t_crit * se
    
    ci_results.append({
        'n': n,
        'mean': mean,
        'lower': mean - moe,
        'upper': mean + moe,
        'width': 2 * moe
    })

# Create DataFrame
ci_df = pd.DataFrame(ci_results)

print("="*60)
print("📏 EFFECT OF SAMPLE SIZE ON CONFIDENCE INTERVALS")
print("="*60)
print()
print(ci_df.to_string(index=False))
print()
print("💡 Key Insight:")
print("   Larger sample size → Narrower confidence interval")
print("   More data → More precision!")
print()
print("="*60)
```

```python
# Visualize CIs for different sample sizes
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: CI ranges
y_positions = range(len(ci_df))
for i, row in ci_df.iterrows():
    axes[0].plot([row['lower'], row['upper']], [i, i], 'o-', linewidth=3, markersize=8)
    axes[0].plot(row['mean'], i, 'ro', markersize=10)

axes[0].axvline(population_mean, color='green', linestyle='--', linewidth=2, 
                label=f'True Mean: {population_mean:.0f}g')
axes[0].set_yticks(y_positions)
axes[0].set_yticklabels([f'n={n}' for n in ci_df['n']])
axes[0].set_xlabel('Body Mass (g)', fontsize=12)
axes[0].set_ylabel('Sample Size', fontsize=12)
axes[0].set_title('95% Confidence Intervals by Sample Size', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(axis='x', alpha=0.3)

# Plot 2: CI width vs sample size
axes[1].plot(ci_df['n'], ci_df['width'], 'o-', linewidth=3, markersize=10, color='#FF6B6B')
axes[1].set_xlabel('Sample Size', fontsize=12)
axes[1].set_ylabel('CI Width (g)', fontsize=12)
axes[1].set_title('Confidence Interval Width vs Sample Size', fontsize=14, fontweight='bold')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("📉 Notice: As sample size increases, CI width decreases!")
```

## Example 5: Comparing Species (Inference in Action)

**Question:** Are Adelie penguins lighter than Gentoo penguins on average?

```python
# Get data for two species
adelie = penguins[penguins['species'] == 'Adelie']['body_mass_g']
gentoo = penguins[penguins['species'] == 'Gentoo']['body_mass_g']

print("="*60)
print("🐧 COMPARING PENGUIN SPECIES")
print("="*60)
print()
print(f"Adelie Penguins (n={len(adelie)}):")
print(f"   Mean: {adelie.mean():.2f} g")
print(f"   Std: {adelie.std():.2f} g")
print()
print(f"Gentoo Penguins (n={len(gentoo)}):")
print(f"   Mean: {gentoo.mean():.2f} g")
print(f"   Std: {gentoo.std():.2f} g")
print()
print(f"Difference: {gentoo.mean() - adelie.mean():.2f} g")
print(f"Gentoo penguins are {gentoo.mean() - adelie.mean():.0f}g heavier on average!")
print()

# Calculate 95% CI for the difference
diff_mean = gentoo.mean() - adelie.mean()
diff_se = np.sqrt((adelie.std()**2 / len(adelie)) + (gentoo.std()**2 / len(gentoo)))
diff_moe = 1.96 * diff_se  # Using z-score for large samples

print(f"95% CI for difference: [{diff_mean - diff_moe:.0f}g, {diff_mean + diff_moe:.0f}g]")
print()
print("="*60)
```

```python
# Visualize comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot comparison
data_to_plot = [adelie, gentoo]
bp = axes[0].boxplot(data_to_plot, labels=['Adelie', 'Gentoo'], patch_artist=True)
bp['boxes'][0].set_facecolor('#FF6B6B')
bp['boxes'][1].set_facecolor('#4ECDC4')
for box in bp['boxes']:
    box.set_alpha(0.7)
axes[0].set_ylabel('Body Mass (g)', fontsize=12)
axes[0].set_title('Body Mass Comparison', fontsize=14, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)

# Histogram comparison
axes[1].hist(adelie, bins=20, alpha=0.7, label='Adelie', color='#FF6B6B', edgecolor='black')
axes[1].hist(gentoo, bins=20, alpha=0.7, label='Gentoo', color='#4ECDC4', edgecolor='black')
axes[1].set_xlabel('Body Mass (g)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_title('Distribution Comparison', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
print("="*60)
print("🎓 KEY TAKEAWAYS: INFERENTIAL STATISTICS")
print("="*60)
print()
print("1. 🌍 POPULATION vs SAMPLE:")
print("   - Population: ALL individuals (often impossible to measure)")
print("   - Sample: A subset we actually measure")
print("   - Goal: Use sample to infer about population")
print()
print("2. 🎯 CENTRAL LIMIT THEOREM:")
print("   - Sample means form a normal distribution")
print("   - Centered around the true population mean")
print("   - Works even if original data isn't normal!")
print()
print("3. 📊 CONFIDENCE INTERVALS:")
print("   - Range of plausible values for population parameter")
print("   - 95% CI: We're 95% confident true value is in this range")
print("   - Larger sample → Narrower CI → More precision")
print()
print("4. 📏 STANDARD ERROR:")
print("   - SE = σ / √n")
print("   - Measures variability of sample means")
print("   - Smaller SE = More reliable estimate")
print()
print("5. 🎲 MARGIN OF ERROR:")
print("   - MOE = t* × SE")
print("   - How much our estimate might be 'off'")
print("   - CI = Sample Mean ± MOE")
print()
print("="*60)
```
