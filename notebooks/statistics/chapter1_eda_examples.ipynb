# Chapter 1: Exploratory Data Analysis (EDA) - Practical Examples

This notebook demonstrates EDA techniques using the **Iris Dataset** - a classic dataset containing measurements of iris flowers.

**Dataset Source:** UCI Machine Learning Repository (built into scikit-learn)

## What You'll Learn:
- Loading and inspecting data
- Summary statistics
- Data visualization
- Finding patterns and outliers

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Set style for better-looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print("✅ Libraries imported successfully!")
```

## Step 1: Load the Data

The Iris dataset contains:
- **150 samples** of iris flowers
- **4 features**: sepal length, sepal width, petal length, petal width (all in cm)
- **3 species**: Setosa, Versicolor, Virginica

```python
# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species_name'] = df['species'].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})

# Display first few rows
print("First 5 rows of the dataset:")
df.head()
```

## Step 2: Basic Data Inspection

Let's understand what we're working with!

```python
# Dataset shape
print(f"📊 Dataset Shape: {df.shape}")
print(f"   - {df.shape[0]} rows (samples)")
print(f"   - {df.shape[1]} columns (features)\n")

# Data types
print("📋 Data Types:")
print(df.dtypes)
print()

# Check for missing values
print("🔍 Missing Values:")
print(df.isnull().sum())
print()

# Basic info
print("ℹ️ Dataset Info:")
df.info()
```

## Step 3: Summary Statistics

Get the big picture with descriptive statistics!

```python
# Summary statistics for numerical columns
print("📈 Summary Statistics:\n")
df.describe()
```

```python
# Count of each species
print("🌸 Species Distribution:\n")
print(df['species_name'].value_counts())
print()

# Visualize species distribution
plt.figure(figsize=(8, 5))
df['species_name'].value_counts().plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Distribution of Iris Species', fontsize=16, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Step 4: Univariate Analysis

Analyze each feature individually

```python
# Histograms for all numerical features
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Iris Features', fontsize=16, fontweight='bold')

features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

for idx, (feature, color) in enumerate(zip(features, colors)):
    ax = axes[idx // 2, idx % 2]
    ax.hist(df[feature], bins=20, color=color, alpha=0.7, edgecolor='black')
    ax.set_title(feature, fontsize=12, fontweight='bold')
    ax.set_xlabel('Value (cm)', fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Box plots to detect outliers
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Box Plots - Detecting Outliers', fontsize=16, fontweight='bold')

for idx, (feature, color) in enumerate(zip(features, colors)):
    ax = axes[idx // 2, idx % 2]
    bp = ax.boxplot(df[feature], patch_artist=True, vert=True)
    bp['boxes'][0].set_facecolor(color)
    bp['boxes'][0].set_alpha(0.7)
    ax.set_title(feature, fontsize=12, fontweight='bold')
    ax.set_ylabel('Value (cm)', fontsize=10)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("💡 Tip: Points outside the 'whiskers' are potential outliers!")
```

## Step 5: Bivariate Analysis

Explore relationships between features

```python
# Scatter plot: Petal Length vs Petal Width
plt.figure(figsize=(10, 6))
colors_map = {'Setosa': '#FF6B6B', 'Versicolor': '#4ECDC4', 'Virginica': '#45B7D1'}

for species in df['species_name'].unique():
    subset = df[df['species_name'] == species]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'], 
                label=species, alpha=0.7, s=100, color=colors_map[species])

plt.title('Petal Length vs Petal Width by Species', fontsize=16, fontweight='bold')
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("🔍 Observation: Different species form distinct clusters!")
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation = df[features].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Iris Features', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n📊 Correlation Insights:")
print("- Strong positive correlation (close to 1): Features move together")
print("- Weak/no correlation (close to 0): Features are independent")
print("- Negative correlation (close to -1): Features move in opposite directions")
```

## Step 6: Multivariate Analysis

Analyze multiple features together

```python
# Pair plot - all features at once!
sns.pairplot(df, hue='species_name', palette={'Setosa': '#FF6B6B', 'Versicolor': '#4ECDC4', 'Virginica': '#45B7D1'},
             diag_kind='kde', height=2.5)
plt.suptitle('Pair Plot - All Features by Species', y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("🎨 This shows ALL relationships at once!")
print("   - Diagonal: Distribution of each feature")
print("   - Off-diagonal: Scatter plots between features")
```

## Step 7: Key Findings Summary

```python
print("="*60)
print("🎯 KEY FINDINGS FROM EDA")
print("="*60)
print()
print("1. 📊 Dataset Overview:")
print(f"   - {df.shape[0]} samples, {df.shape[1]} features")
print(f"   - 3 species: {', '.join(df['species_name'].unique())}")
print(f"   - No missing values ✅")
print()
print("2. 📈 Feature Distributions:")
print(f"   - Petal length: {df['petal length (cm)'].min():.1f} - {df['petal length (cm)'].max():.1f} cm")
print(f"   - Petal width: {df['petal width (cm)'].min():.1f} - {df['petal width (cm)'].max():.1f} cm")
print(f"   - Sepal length: {df['sepal length (cm)'].min():.1f} - {df['sepal length (cm)'].max():.1f} cm")
print(f"   - Sepal width: {df['sepal width (cm)'].min():.1f} - {df['sepal width (cm)'].max():.1f} cm")
print()
print("3. 🔍 Correlations:")
strongest_corr = correlation.abs().unstack().sort_values(ascending=False)
strongest_corr = strongest_corr[strongest_corr < 1.0].head(1)
print(f"   - Strongest correlation: {strongest_corr.index[0][0]} & {strongest_corr.index[0][1]}")
print(f"     Correlation coefficient: {strongest_corr.values[0]:.3f}")
print()
print("4. 🌸 Species Patterns:")
print("   - Setosa has smallest petals")
print("   - Virginica has largest petals")
print("   - Species form distinct clusters")
print()
print("5. 💡 Outliers:")
print("   - Few outliers detected in sepal width")
print("   - Most data points are well-behaved")
print()
print("="*60)
print("✅ EDA Complete! Ready for modeling or hypothesis testing.")
print("="*60)
```
