```python
# Load tips dataset
tips = sns.load_dataset('tips')

print("📊 Dataset Shape:", tips.shape)
print("\nFirst 5 rows:")
tips.head()
```

## Example 1: One-Sample t-Test

### Question: Is the average tip $3.00?

**Scenario:** A restaurant manager claims that the average tip is $3.00. Let's test this claim!

### The 5-Step Process:

```python
print("="*60)
print("🎯 HYPOTHESIS TEST: Is the average tip $3.00?")
print("="*60)
print()

# Step 1: State the hypotheses
print("📝 STEP 1: State the Hypotheses")
print("   H₀ (Null): μ = $3.00 (average tip is $3.00)")
print("   H₁ (Alternative): μ ≠ $3.00 (average tip is NOT $3.00)")
print()

# Step 2: Set significance level
alpha = 0.05
print("📊 STEP 2: Set Significance Level")
print(f"   α = {alpha} (5% chance of Type I error)")
print()

# Step 3: Collect and analyze data
print("📈 STEP 3: Collect and Analyze Data")
sample_mean = tips['tip'].mean()
sample_std = tips['tip'].std()
sample_size = len(tips)
claimed_mean = 3.00

print(f"   Sample size: {sample_size}")
print(f"   Sample mean: ${sample_mean:.2f}")
print(f"   Sample std: ${sample_std:.2f}")
print(f"   Claimed mean: ${claimed_mean:.2f}")
print()

# Visualize the data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(tips['tip'], bins=30, color='#4ECDC4', alpha=0.7, edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f'Sample Mean: ${sample_mean:.2f}')
plt.axvline(claimed_mean, color='orange', linestyle='--', linewidth=2, label=f'Claimed Mean: ${claimed_mean:.2f}')
plt.xlabel('Tip Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Tips', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot(tips['tip'], vert=True, patch_artist=True,
            boxprops=dict(facecolor='#4ECDC4', alpha=0.7))
plt.axhline(claimed_mean, color='orange', linestyle='--', linewidth=2, label=f'Claimed: ${claimed_mean:.2f}')
plt.ylabel('Tip Amount ($)', fontsize=12)
plt.title('Box Plot of Tips', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Step 4: Calculate test statistic and p-value
print("🔬 STEP 4: Calculate Test Statistic and p-value")
print()

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(tips['tip'], claimed_mean)

print(f"   t-statistic: {t_statistic:.4f}")
print(f"   p-value: {p_value:.4f}")
print()
print("   💡 What does this mean?")
print(f"      - If H₀ is true, there's a {p_value*100:.2f}% chance of seeing")
print(f"        a sample mean this far from ${claimed_mean:.2f}")
print()

# Step 5: Make decision
print("⚖️ STEP 5: Make Decision")
print()
if p_value < alpha:
    print(f"   ✅ REJECT H₀ (p-value {p_value:.4f} < α {alpha})")
    print(f"   📊 Conclusion: The average tip is NOT $3.00")
    print(f"   📈 The true average is likely ${sample_mean:.2f}")
else:
    print(f"   ❌ FAIL TO REJECT H₀ (p-value {p_value:.4f} >= α {alpha})")
    print(f"   📊 Conclusion: Not enough evidence to say average tip ≠ $3.00")
    print(f"   📈 The claim of $3.00 could be true")

print()
print("="*60)
```

## Example 2: Two-Sample t-Test

### Question: Do smokers tip differently than non-smokers?

**Scenario:** We want to know if smoking status affects tipping behavior.

```python
print("="*60)
print("🎯 HYPOTHESIS TEST: Do smokers tip differently?")
print("="*60)
print()

# Step 1: State hypotheses
print("📝 STEP 1: State the Hypotheses")
print("   H₀: μ_smokers = μ_non-smokers (no difference in tips)")
print("   H₁: μ_smokers ≠ μ_non-smokers (tips differ by smoking status)")
print()

# Step 2: Significance level
print("📊 STEP 2: Set Significance Level")
print(f"   α = {alpha}")
print()

# Step 3: Analyze data
print("📈 STEP 3: Collect and Analyze Data")
smokers = tips[tips['smoker'] == 'Yes']['tip']
non_smokers = tips[tips['smoker'] == 'No']['tip']

print(f"   Smokers: n={len(smokers)}, mean=${smokers.mean():.2f}, std=${smokers.std():.2f}")
print(f"   Non-smokers: n={len(non_smokers)}, mean=${non_smokers.mean():.2f}, std=${non_smokers.std():.2f}")
print(f"   Difference: ${smokers.mean() - non_smokers.mean():.2f}")
print()

# Visualize
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.hist(smokers, bins=20, alpha=0.7, label='Smokers', color='#FF6B6B', edgecolor='black')
plt.hist(non_smokers, bins=20, alpha=0.7, label='Non-smokers', color='#4ECDC4', edgecolor='black')
plt.xlabel('Tip Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Tip Distribution by Smoking Status', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
data_to_plot = [smokers, non_smokers]
bp = plt.boxplot(data_to_plot, labels=['Smokers', 'Non-smokers'], patch_artist=True)
bp['boxes'][0].set_facecolor('#FF6B6B')
bp['boxes'][1].set_facecolor('#4ECDC4')
for box in bp['boxes']:
    box.set_alpha(0.7)
plt.ylabel('Tip Amount ($)', fontsize=12)
plt.title('Box Plot Comparison', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Step 4: Calculate test statistic
print("🔬 STEP 4: Calculate Test Statistic and p-value")
print()

# Perform independent two-sample t-test
t_stat, p_val = stats.ttest_ind(smokers, non_smokers)

print(f"   t-statistic: {t_stat:.4f}")
print(f"   p-value: {p_val:.4f}")
print()

# Step 5: Decision
print("⚖️ STEP 5: Make Decision")
print()
if p_val < alpha:
    print(f"   ✅ REJECT H₀ (p-value {p_val:.4f} < α {alpha})")
    print(f"   📊 Conclusion: Smokers and non-smokers tip DIFFERENTLY")
    if smokers.mean() > non_smokers.mean():
        print(f"   📈 Smokers tip MORE on average (${smokers.mean():.2f} vs ${non_smokers.mean():.2f})")
    else:
        print(f"   📉 Smokers tip LESS on average (${smokers.mean():.2f} vs ${non_smokers.mean():.2f})")
else:
    print(f"   ❌ FAIL TO REJECT H₀ (p-value {p_val:.4f} >= α {alpha})")
    print(f"   📊 Conclusion: No significant difference in tipping")
    print(f"   📈 Smoking status doesn't affect tips")

print()
print("="*60)
```

## Example 3: Proportion Test

### Question: Do more than 50% of customers dine on weekends?

**Scenario:** Testing if weekend dining is more popular than weekday dining.

```python
print("="*60)
print("🎯 HYPOTHESIS TEST: Weekend vs Weekday Dining")
print("="*60)
print()

# Classify as weekend or weekday
tips['is_weekend'] = tips['day'].isin(['Sat', 'Sun'])
weekend_count = tips['is_weekend'].sum()
total_count = len(tips)
weekend_proportion = weekend_count / total_count

print("📝 STEP 1: State the Hypotheses")
print("   H₀: p = 0.50 (50% dine on weekends)")
print("   H₁: p > 0.50 (more than 50% dine on weekends)")
print()

print("📊 STEP 2: Set Significance Level")
print(f"   α = {alpha}")
print()

print("📈 STEP 3: Collect and Analyze Data")
print(f"   Total customers: {total_count}")
print(f"   Weekend customers: {weekend_count}")
print(f"   Weekend proportion: {weekend_proportion:.2%}")
print()

# Visualize
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
counts = tips['is_weekend'].value_counts()
plt.bar(['Weekday', 'Weekend'], [counts[False], counts[True]], 
        color=['#4ECDC4', '#FF6B6B'], alpha=0.7, edgecolor='black')
plt.ylabel('Number of Customers', fontsize=12)
plt.title('Weekday vs Weekend Dining', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
plt.pie([counts[False], counts[True]], labels=['Weekday', 'Weekend'], 
        autopct='%1.1f%%', colors=['#4ECDC4', '#FF6B6B'], startangle=90)
plt.title('Proportion of Dining Days', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Step 4: Calculate test statistic
print("🔬 STEP 4: Calculate Test Statistic and p-value")
print()

# One-sample proportion test (using binomial test)
from statsmodels.stats.proportion import proportions_ztest

# z-test for proportions
z_stat, p_val_prop = proportions_ztest(weekend_count, total_count, 0.50, alternative='larger')

print(f"   z-statistic: {z_stat:.4f}")
print(f"   p-value: {p_val_prop:.4f}")
print()

# Step 5: Decision
print("⚖️ STEP 5: Make Decision")
print()
if p_val_prop < alpha:
    print(f"   ✅ REJECT H₀ (p-value {p_val_prop:.4f} < α {alpha})")
    print(f"   📊 Conclusion: MORE than 50% dine on weekends")
    print(f"   📈 Actual proportion: {weekend_proportion:.2%}")
else:
    print(f"   ❌ FAIL TO REJECT H₀ (p-value {p_val_prop:.4f} >= α {alpha})")
    print(f"   📊 Conclusion: Not enough evidence that >50% dine on weekends")
    print(f"   📈 Proportion could be 50% or less")

print()
print("="*60)
```

## Understanding p-values Visually

```python
# Visualize p-value concept
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Generate normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

# Plot 1: Small p-value (reject H0)
axes[0].plot(x, y, 'b-', linewidth=2)
axes[0].fill_between(x[x > 2.5], 0, y[x > 2.5], alpha=0.3, color='red')
axes[0].fill_between(x[x < -2.5], 0, y[x < -2.5], alpha=0.3, color='red')
axes[0].axvline(2.5, color='red', linestyle='--', linewidth=2)
axes[0].set_title('Small p-value (< 0.05)\nREJECT H₀', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Test Statistic', fontsize=10)
axes[0].set_ylabel('Probability', fontsize=10)
axes[0].text(2.5, 0.2, 'Observed\nvalue', ha='center', fontsize=9)
axes[0].grid(alpha=0.3)

# Plot 2: Large p-value (fail to reject H0)
axes[1].plot(x, y, 'b-', linewidth=2)
axes[1].fill_between(x[x > 1], 0, y[x > 1], alpha=0.3, color='green')
axes[1].fill_between(x[x < -1], 0, y[x < -1], alpha=0.3, color='green')
axes[1].axvline(1, color='green', linestyle='--', linewidth=2)
axes[1].set_title('Large p-value (> 0.05)\nFAIL TO REJECT H₀', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Test Statistic', fontsize=10)
axes[1].text(1, 0.2, 'Observed\nvalue', ha='center', fontsize=9)
axes[1].grid(alpha=0.3)

# Plot 3: Critical region
axes[2].plot(x, y, 'b-', linewidth=2)
critical_value = stats.norm.ppf(0.975)  # Two-tailed, α=0.05
axes[2].fill_between(x[x > critical_value], 0, y[x > critical_value], alpha=0.3, color='red', label='Rejection Region')
axes[2].fill_between(x[x < -critical_value], 0, y[x < -critical_value], alpha=0.3, color='red')
axes[2].fill_between(x[(x > -critical_value) & (x < critical_value)], 0, 
                     y[(x > -critical_value) & (x < critical_value)], alpha=0.3, color='green', label='Acceptance Region')
axes[2].set_title('Critical Regions (α=0.05)', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Test Statistic', fontsize=10)
axes[2].legend(fontsize=9)
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("💡 Understanding p-values:")
print("   - Small p-value (< 0.05): Observed value is RARE under H₀ → Reject H₀")
print("   - Large p-value (> 0.05): Observed value is COMMON under H₀ → Fail to reject H₀")
print("   - Critical region: If test statistic falls here, reject H₀")
```

## 🎓 Key Takeaways:

### The 5-Step Process:
1. **State hypotheses** (H₀ and H₁)
2. **Set significance level** (α = 0.05)
3. **Collect and analyze data**
4. **Calculate test statistic and p-value**
5. **Make decision** (reject or fail to reject H₀)

### Decision Rules:
- **p-value < α**: Reject H₀ (result is statistically significant)
- **p-value ≥ α**: Fail to reject H₀ (not enough evidence)

### Types of Tests:
- **One-sample**: Compare sample to a claimed value
- **Two-sample**: Compare two groups
- **Proportion**: Test percentages/proportions
